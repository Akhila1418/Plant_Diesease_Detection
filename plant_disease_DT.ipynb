```python
import os
import cv2
import numpy as np
from skimage.feature import hog
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```python
data_dir = "D:\\PLANT_DISEASE\\archive (7)\\New Plant Diseases Dataset(Augmented)\\New Plant Diseases Dataset(Augmented)\\train_new" 
```

```python
feature_vectors = []
labels = []
```

```python
orientations = 9
pixels_per_cell = (8, 8)
cells_per_block = (2, 2)
```

```python
for class_label in os.listdir(data_dir):
    class_dir = os.path.join(data_dir, class_label)
    
    if not os.path.isdir(class_dir):
        continue

    for image_file in os.listdir(class_dir):
        image_path = os.path.join(class_dir, image_file)
        
        image = cv2.imread(image_path)
        
        image = cv2.resize(image, (128, 128))
        
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        features = hog(image, orientations=orientations, pixels_per_cell=pixels_per_cell,
                       cells_per_block=cells_per_block, visualize=False, block_norm='L2-Hys')
        
        feature_vectors.append(features)
        labels.append(class_label)
```

```python
feature_vectors
```

```python
labels
```

```python
X = np.array(feature_vectors)
y = np.array(labels)
```

```python
print(X)
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
y
```

```python
dt_model = DecisionTreeClassifier(
    max_depth=100,  
    min_samples_split=2,  
    min_samples_leaf=1,  
    criterion="gini"  
)
```

```python
dt_model.fit(X, y)
```

```python
from joblib import dump

dump(dt_model, 'DT_PLANT_DISEASE.joblib')
```
